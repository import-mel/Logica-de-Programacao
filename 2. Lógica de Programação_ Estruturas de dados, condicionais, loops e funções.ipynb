### **SET Conjuntos**

> Um set (conjunto) é uma coleção não ordenada e mutável de elementos únicos. Essa estrutura de dados é semelhante a conjuntos na matemática.

**Características importantes:**

*   Um set não pode conter elementos duplicados.
*   Sets são mutáveis, o que significa que você pode adicionar e remover elementos após a criação do set.
* Sets suportam operações de conjuntos, como união, interseção e diferença. Essas operações podem ser realizadas usando métodos ou operadores.
* Os elementos em um set não têm uma ordem específica, e você não pode acessá-los por índices, como em listas ou tuplas.




````python
# Criando um conjunto

conjunto = {1, 2, 3, 4, 5, 6}

type(conjunto)
````

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1.union(set2)      # União: {1, 2, 3, 4, 5}
intersection_set = set1.intersection(set2)  # Interseção: {3}
````

````python
print(union_set)
print(intersection_set)
````

### **Tabela-Verdade**

Para entendermos qual será o resultado obtido entre uma expressão relacional utilizando também operadores lógicos para efetuar várias comparações em uma mesma expressão, utilizamos a tabela-verdade.

> Uma tabela-verdade é uma representação tabular de todas as combinações possíveis de valores lógicos para um conjunto específico de proposições ou variáveis booleanas. Ela é usada na lógica matemática para analisar o comportamento de expressões lógicas e determinar o valor lógico resultante para cada combinação de entradas. </br> </br> As proposições ou variáveis booleanas podem ter apenas dois valores possíveis: verdadeiro (T) ou falso (F). **A tabela-verdade lista todas as combinações possíveis desses valores para as variáveis envolvidas em uma expressão lógica** e, em seguida, especifica o valor lógico resultante da expressão para cada combinação.

Para cada operador lógico há uma tabela-verdade.

### **Operação de Conjunção (AND)**

A operação de conjunção representa a relação entre duas proposições (ou expressões booleanas) em que a afirmação composta é verdadeira apenas se **ambas as proposições individuais forem verdadeiras**.

Ou seja, basta uma expressão da relação ser falsa para a relação inteira ser falsa também.

EXEMPLO:

*   O livro é azul **E** possui 200 páginas.
*   O número 9 é maior que 7 **E** menor que 10.
*   Se chover **E** relampejar, eu fico em casa. (Quando eu fico em casa?)

**Legenda: V - Verdadeiro e F - Falso**

<p align="center">
  <img src="https://static.todamateria.com.br/upload/lo/gi/logicaconjuncao-0.jpg?auto_optimize=low" alt="Descrição da imagem" width="300"/>
</p>

**Obs.: o operador E/AND é representado na matemática pelo simbolo ^.**







````python
proposicao_a = 1 >= 1 # é TRUE pois 1 é igual ou maior que 1
proposicao_b = 2 + 4 < 9 # também é verdadeiro pois 2 + 4 é igual a 6, e 6 é menor que 9
proposicao_c = 8 != 8 # será falso pois 8 é igual a 8 e não diferente dele mesmo

print(proposicao_a and proposicao_b and proposicao_c) # como a proposiçao c é falsa logo toda a expressão é falsa também
print(proposicao_a and proposicao_b)
````

### **Operação de Disjunção (OR)**

> Disjunção: Em termos mais gerais, é utilizada para descrever a condição de coisas que estão desconectadas, distintas ou que não compartilham uma relação direta.

A operação de disjunção é uma operação lógica que representa a relação entre duas proposições (ou expressões booleanas) em que a afirmação composta **é considerada verdadeira se pelo menos uma das proposições individuais for verdadeira**.

Ou seja, diferente da operação de conjunção, a de disjunção só precisa que uma das expressão da relação ser verdadeira para todo o resto da expressão ser também.

EXEMPLO:

*   A capa do livro é azul **OU** vermelha.
*   O número 9 é maior que 10 **OU** menor que 10.
*   Se chover **ou** relampejar, eu fico em casa. (Quando eu fico em casa?)

<p align="center">
  
</p>

**Obs.: o operador OU/OR é representado na matemática pelo simbolo ∨.**


````python
proposicao_a = 1 >= 1 # é TRUE pois 1 é igual ou maior que 1
proposicao_b = 2 + 4 < 9 # também é verdadeiro pois 2 + 4 é igual a 6, e 6 é menor que 9
proposicao_c = 8 != 8 # será falso pois 8 é igual a 8 e não diferente dele mesmo

print(proposicao_a or proposicao_b or proposicao_c)
print(proposicao_a or proposicao_c) # mesmo a proposiçao c sendo falsa, a proposicao a é verdadeira logo toda a expressao será verdadeira
````

````python
# sendo todos as proposições falsas, a expressão também será falsa

proposicao_d = 1 < -1
proposicao_e = 3 <= 0

print(proposicao_d or proposicao_e)
````

````python
# Também podemos combinar numa mesma expressão os operadores OR e AND

conjuncao = 3 > 9 and 3 < 8 #FALSE
disjuncao = 49 < 100 or 34000 != 340000 #VERDADEIRO

print(conjuncao and disjuncao) #FALSE AND TRUE
print(conjuncao or disjuncao)  #FALSE OR TRUE
````

````python
# se escrevermos isso direto no print() então precisamos organizar as expressões
print((3 > 9 and 3 < 8) and (49 < 100 or 34000 != 340000))
print(3 > 9 and 3 < 8 and 49 < 100 or 34000 != 340000) #observe que aqui ele deu como true, porque seguir a ordem dos elementos, se colocamos cada expressao num parenteses ele funciona igual a regra matemática

print((3 > 9 and 3 < 8) or (49 < 100 or 34000 != 340000))
print(3 > 9 and 3 < 8 or 49 < 100 or 34000 != 340000)
````

### **Operação de Negação (NOT)**

A operação de negação é uma forma de dizer o oposto de uma afirmação, **é uma maneira de inverter o sentido de uma declaração.**

Se algo é verdadeiro, a negação diz que é falso, e se algo é falso, a negação diz que é verdadeiro. Por exemplo, se afirmamos "Está chovendo" e aplicamos a negação, obtemos "Não está chovendo".

<p align="center">
  <img src="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSpSnDM4xHImCdsIv7EDpm4kvNmiZIGCj-pgw&usqp=CAU" alt="Descrição da imagem" width="300"/>
</p>

**Obs.: o operador não/NOT é representado na matemática pelo simbolo ~.**











````python
proposicao_f = 3 == 3 #aqui será verdadeiro, pois 3 é igual a ele mesmo

#então se usarmos o operador de negacao, então ele irá inverter o valor lógico dessa expressão, o que é V vira F
not(proposicao_f)
````

````python
# Utilizando em expressões maiores:

conjuncao = 3 > 9 and 3 < 8 #FALSE
disjuncao = 49 < 100 or 34000 != 340000 #VERDADEIRO

print(not(conjuncao and disjuncao)) #FALSE AND TRUE = FALSE, porem not(false) = true
print(not(conjuncao or disjuncao))  #FALSE OR TRUE = TRUE, porem not(true) = false
````

**Observação**: caso numa mesma expressão haja todos os operadores, então para resolve-la é necessário seguir uma ordem de prioridade:

*   **Primeiro:** parênteses mais internos;
*   **Segundo:** operadores aritméticos;
*   **Terceiro:** operadores relacionais;
*   **Quarto:** operadores lógicos.


# Estruturas de Controle

Em programação, as "estruturas de controle" referem-se a mecanismos que permitem **controlar o fluxo de execução de um programa**. Elas são fundamentais para determinar como as instruções em um programa são executadas com base em certas condições ou iterações. Existem três tipos principais:

*     Estrutura Sequencial
*     Estrutura Condicional
*     Estrutura de Repetição

Em outras palavras, estrutura de controle refere-se à ordem em que instruções, expressões e chamadas de função são executadas.

### **Estrutura Sequencial**

Ao construirmos um algoritmo nessa estrutura, o programa irá **executar as intruções de forma linear**, ou seja, de cima para baixo e da esquerda para a direita (na mesma ordem que foram escritas).

#### **Estrutura Condicional Simples**

A estrutura condicional simples expõe que uma instrução ou bloco de instruções só sejam executados se determinada condição for verdadeira. Caso a condição for falsa, então o algoritmo finaliza sem executar nenhum comando.

#### **Estrutura Condicional Composta**

A estrutura condicional composta expõe que uma instrução ou bloco de instruções só sejam executados se determinada condição for satisfeita. Caso essa condição seja falsa, é executada uma outra instrução ou um bloco de instruções diferente.

> **Ou seja, a estrutura condicional composta executa um bloco de instruções se a condição for verdadeira e outro se a condição for falsa.**

Em pseudocódigo utilizamos o SE caso a condição for verdadeira e o SENÃO caso a condição for falsa. O SENÃO em python é representado pelo ELSE.

````python
# Utilizaremos o mesmo código do exemplo da condicional simples, mas agora implementaremos o else para dizer que o aluno foi reprovado

nome_aluno = input("Digite o nome do aluno: ")

primeiro_semestre = float(input("Digite a nota do aluno no primeiro semestre do ano: "))
segundo_semestre = float(input("Digite a nota do aluno no segundo semestre do ano: "))
terceiro_semestre = float(input("Digite a nota do aluno no terceiro semestre do ano: "))
quarto_semestre = float(input("Digite a nota do aluno no quarto semestre do ano: "))

media = (primeiro_semestre + segundo_semestre + terceiro_semestre + quarto_semestre) / 4

if media >= 6:
    print(f"A nota final de {nome_aluno} foi: {media}")
    print("Aluno Aprovado!")
else :
    print(f"A nota final de {nome_aluno} foi {media}, ficando abaixo da média exigida.")
    print("Aluno Reprovado!")
````

Observe que agora o código apresenta uma resposta caso o aluno for reprovado, diferente na condicional simples onde não exibia nada.

#### **Estrutura Condicional Encadeada (ou aninhada)**

Essa estrutura refere-se à organização de **condicionais dentro de outras condicionais**. Isso ocorre quando você tem uma estrutura condicional dentro de outra, criando uma hierarquia ou encadeamento de decisões.

A ideia é que, se a primeira condição não for satisfeita, você pode ter uma segunda condição para verificar. Se a segunda condição também não for satisfeita, pode haver uma terceira, e assim por diante. **Isso permite lidar com casos mais complexos onde há múltiplas condições a serem consideradas.**

> Mostraremos duas formas de criar essa estrutura em python, uma utilizando as palavras-chaves que já conhecemos if e else, e a segunda utilizando elif (else if em outras linguagens, significa senão se).


Utilizasemos o mesmo exemplo de antes, mas agora com outros condicionais:
*  caso a media do aluno for igual a 4.5 e menor que 6 mostraremos a mensagem: Aluno está de recupereção!
*  caso a media do aluno for menor  4.5 irá mostrar a mensagem: Aluno Reprovado sem direito a recuperação!
* caso a media do aluno for exatamente 6 irá mostrar a mensagem: Passou Raspando!

````python
nome_aluno = input("Digite o nome do aluno: ")

primeiro_semestre = float(input("Digite a nota do aluno no primeiro semestre do ano: "))
segundo_semestre = float(input("Digite a nota do aluno no segundo semestre do ano: "))
terceiro_semestre = float(input("Digite a nota do aluno no terceiro semestre do ano: "))
quarto_semestre = float(input("Digite a nota do aluno no quarto semestre do ano: "))

media = (primeiro_semestre + segundo_semestre + terceiro_semestre + quarto_semestre) / 4

if media > 6:
    print(f"A nota final de {nome_aluno} foi: {media}")
    print("Aluno Aprovado!")
else :
    if media == 6:
        print(f"A nota final de {nome_aluno} foi: {media}")
        print("Aluno Aprovado, passou raspando!")
    else:
        if media >= 4.5 and media < 6:
            print(f"A nota final de {nome_aluno} foi: {media}")
            print("Aluno Reprovado, mas tem direito a tentar a recuperação!")
        else:
            print(f"A nota final de {nome_aluno} foi {media}, ficando abaixo da média exigida.")
            print("Aluno Reprovado, sem direito a recuperação!")
````

## **Estrutura de Repetição**

> Os laçoes de repetição também são conhecidos por sua tradução em inglês: loops ou looping. Essa estrutura é utilizada para executar um bloco de código várias vezes, com base em uma condição específica, sendo fundamentais para automatizar tarefas repetitivas.

Em python temos duas formas de construir um laço, a primeira usando **WHILE** e a segunda utilizando **FOR**.

### **COMANDO WHILE**

Consiste em uma estrutura de controle do fluxo de execução que permite repetir diversas vezes um mesmo trecho do algoritmo, porém, **sempre verificando antes de cada execução se é permitido executar o mesmo trecho**.

> Ou seja, antes de executar ele verifica se a condição é verdadeira, e irá repetir o mesmo bloco de código até a condição se tornar falsa.

Em pseudocódigo, o while é representado pela palavra enquanto, podemos pensar da seguinte forma - **ENQUANTO a condição for verdadeira FAÇA**.

````python
# iremos utilizar o mesmo exemplo da tabuada criada na estrutura sequencial, mas agora sem precisar escrever 10 print

numero = int(input("Digite o número inteiro que você quer que exiba a tabuada: "))
contador = 0

while contador <= 10:
    print(f"{numero} x {contador} = {numero * contador}")
    contador += 1 #mesma coisa de contador = contador + 1
````

Observe que se não adicionarmos o trecho **contador += 1** o código será executado para sempre porque o contador não mudará e o loop irá se repetir até ele ser menor ou igual a dez.

Agora podemos fazer algo mais complexo, onde podemos digitar a media da nota de 10 alunos e adicionar isso numa estrutura de dados do tipo dicionário, onde o nome de cada aluno será a chave e sua nota final será o valor.

Além disso, criar três dicionários, um para os alunos aprovados, outro pros alunos em recuperação e outro pros alunos reprovados.

````python
quantidade_alunos = int(input("Digite a quantidade de alunos: "))

alunos_aprovados = {}
alunos_recuperacao = {}
alunos_reprovados = {}

while quantidade_alunos != 0:
    #nome
    nome_aluno = input("Digite o nome do aluno: ")

    #notas
    primeiro_semestre = float(input("Digite a nota do aluno no primeiro semestre do ano: "))
    segundo_semestre = float(input("Digite a nota do aluno no segundo semestre do ano: "))
    terceiro_semestre = float(input("Digite a nota do aluno no terceiro semestre do ano: "))
    quarto_semestre = float(input("Digite a nota do aluno no quarto semestre do ano: "))

    #nota final
    media = (primeiro_semestre + segundo_semestre + terceiro_semestre + quarto_semestre) / 4

    if media > 6:
        print(f"A nota final de {nome_aluno} foi: {media}")
        print("Aluno Aprovado!")
        alunos_aprovados[nome_aluno] = media
    elif media == 6:
        print(f"A nota final de {nome_aluno} foi: {media}")
        print("Aluno Aprovado, passou raspando!")
        alunos_aprovados[nome_aluno] = media
    elif media >= 4.5 and media < 6:
        print(f"A nota final de {nome_aluno} foi: {media}")
        print("Aluno Reprovado, mas tem direito a tentar a recuperação!")
        alunos_recuperacao[nome_aluno] = media
    else:
        print(f"A nota final de {nome_aluno} foi {media}, ficando abaixo da média exigida.")
        print("Aluno Reprovado, sem direito a recuperação!")
        alunos_reprovados[nome_aluno] = media

    quantidade_alunos -= 1 # o mesmo que quantidade_alunos = quantidade_alunos - 1

print("--------------------------------------------------------------------------")
print("ALUNOS APROVADOS:", alunos_aprovados)
print(f"TOTAL: {len(alunos_aprovados)}")
print("--------------------------------------------------------------------------")
print("ALUNOS EM RECUPERAÇÃO:", alunos_recuperacao)
print(f"TOTAL: {len(alunos_recuperacao)}")
print("--------------------------------------------------------------------------")
print("ALUNOS REPTROVADOS:", alunos_reprovados)
print(f"TOTAL: {len(alunos_reprovados)}")
print("--------------------------------------------------------------------------")
````

### **COMANDO FOR**

A estrutura **for** em Python é uma forma de loop que é utilizada para **iterar** sobre uma sequência de elementos. Pode ser uma lista, uma tupla ou um dicionário, por exemplo.

**ITERAR:** é o processo chamado na programação de repetição de uma ou mais ações.

> A diferença principal dessa estrutura e a outra acima (while) é que nessa sabemos quantas vezes o bloco de código terá que ser executado, em while apenas determinavamos uma condição e ele repetia as intruções até que a condição não fosse mais satisfeita.

Em pseudocódigo, for é representado pela palavra para, podemos pensar da seguinte forma caso queiramos percorrer uma lista - **PARA cada elemento dessa lista FAÇA.**

````python
# Primeiro exemplo: vamos determinar quantas vezes queremos que ele execute o mesmo bloco de código

numero = int(input("Digite o número inteiro que você quer que exiba a tabuada: "))

# utilizando a função range
for i in range(1, 11): # vai gerar o intervalos de números de 1 á 10, lembrando que não é 11 porque começa a partir do zero
    print(f"{numero} X {i} = {numero * i}")
````

Perceba que não foi preciso escrever no código **i += 1** como fizemos no WHILE, pois essa estrutura já soma (incrementa) automaticamente.

Em outras linguagens escrevemos o for da seguinte forma:

```
for(i = 0; i <= 10; i += 1)
```

Essa estrutura que está embutida no range, onde definimos uma váriavel, e o for irá executar o bloco de código até a condição **i <= 10** não for mais satisfeita, e o **i += 1** que incrementa a cada rodada para que o loop não se torne infinito.



````python
# Segundo exemplo: Vamos percorrer uma lista

for rodada in [1, 2, 3, 4, 5, 6]:
    print(rodada)
````

````python
dict = {'chave1':1230, 'chave2':[22,453,73.4], 'chave3':['picanha', 'fraldinha', 'alcatra']}

for rodada in dict.values():
    print(rodada)

for rodada in dict:
    print(rodada)
````

````python
# Colocando um intervalo entre os números

for rodada in range(1, 100, 2):
    print(rodada)
````

# O que são blocos de código?

Um bloco de código em programação é um **conjunto de instruções que são agrupadas e executadas juntas como uma unidade lógica**.

Em muitas linguagens de programação, os blocos de código são delimitados por chaves, colchetes ou palavras-chave específicas (como if, while, for). Esses blocos podem conter uma ou mais instruções e também podem estar aninhados dentro de outros blocos de código.

Os blocos de código são definidos por identação, como fizemos nos exemplos de estruturas de controle, cada um daqueles if, elif, else, while e for são blocos de códigos que executam determinadas instruções.

# Funções

Já utilizamos	algumas	funções	já	prontas	do	Python,	como	o	print(),	input(),	format()	e	type(), mas também	podemos	criar	nossas	próprias	funções.

> Na programação, funções são blocos de código que realizam determinadas **tarefas que normalmente precisam ser executadas diversas vezes dentro de uma aplicação**. </br> </br> Quando surge essa necessidade, para que várias instruções não precisem ser repetidas (escrever diversar linhas de código iguais), elas são agrupadas em uma função, à qual é dado um nome e que poderá ser chamada/executada em diferentes partes do programa.</br> </br> **Ou seja, uma função é um bloco de código reutilizável que realiza uma tarefa específica.**

As funções são essenciais para a modularização e organização de código, pois permitem dividir um programa em partes menores e mais gerenciáveis, sendo são uma ferramenta poderosa para criar código mais legível, modular e fácil de dar manutenção.







## **Criando uma função**

Uma função geralmente tem um nome que a identifica, uma lista de parâmetros (ou argumentos) que podem ser passados para ela, e um corpo, que é o conjunto de instruções que define o que a função faz quando é chamada.

**Obs.:** Igual quando utilizamos a função input, onde o programa solicita uma entrada de dados, podemos fazer a mesma coisa com uma função, onde quando for declarada devemos definir o dado (chamado de **parâmetro**) que ela utilizará na execução das instruções definidas dentro dela. Porém, não é obrigatório que todas as funções sejam construidas para receber parâmetros.

> Assim como quando utilizamos a função **input** para solicitar uma entrada de dados ao programa, quando declaramos uma função, também podemos definir os dados que ela irá receber, conhecidos como **parâmetros**, para utilizar na execução das instruções dentro dela. No entanto, é importante ressaltar que nem todas as funções precisam necessariamente receber parâmetros para funcionar corretamente.

````python
# Em python usamos a palavra chave def nome_da_função(): para iniciarmos uma função

#função que não recebe entrada de parametros
def mensagem_feliz():
    print("Ninguém sabe como tá a mente do palhaço") #quando chamarmos essa função pelo seu nome, o sistema irá imprimir essa mensagem

mensagem_feliz() #nunca esqueça dos parenteses, eles que definem que isso é uma função e que ela está sendo chamada
````

````python
# Função que precisa receber parametros

def mensagem(msg):
    print(msg)

mensagem("Saudam a mandioca!!!")
````

````python
# Função que recebe o valor da tabuada que o usuário quer e até que numero ele quer que seja multiplicado, ou seja, dois parametros

# Para a função receber dois parametros, separamos cada parametro com uma virgula
def tabuada(numero_tabuada, valor):
    for i in range(valor+1):
        print(f"{numero_tabuada} X {i} = {numero_tabuada * i}")

tabuada(12, 20)
````

Veja que a função facilita e muito o trabalho de um programador, pois criando apenas uma vez a estrutura, podemos utiliza-la a qualquer momento no código apenas chamando o nome dela em vez de escrever o mesmo bloco de código novamente.

# Escopo Global e Local

O escopo em programação se refere à visibilidade e acessibilidade de variáveis em diferentes partes do código.

## **Escopo Global**

* Variáveis definidas fora de todos os blocos de código têm escopo global.
* Elas podem ser acessadas de qualquer lugar do código, dentro ou fora das funções.


````python
# EXEMPLO DE VÁRIAVEL GLOBAL:

x = 1

if (x == 1):
    print("oi")

x
````

* Elas só podem ser acessadas dentro dessa função.

* O escopo local é limitado ao bloco de código onde a variável é declarada.
