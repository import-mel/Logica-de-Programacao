## Tipos Primitivos

Os tipos primitivos (também chamados de tipos básicos) são categorias fundamentais de dados. É a partir deles que estruturas de dados mais complexas são construídas.

Podemos dividir os tipos de dados que um computador pode manipular em quatro tipos:


1.   **Inteiro (numérico):** toda informação numérica que pertence ao conjunto dos números inteiros (negativo, positivo ou nulo);
2.   **Números Reais (números decimais ou ponto flutuante):** negrito toda informação numérica que pertece ao conjunto dos números reais;
3. **Caracteres (cadeia de caracteres):** letras, frases, textos, toda informação que pertença ao conjunto dos caracteres alfanuméricos (0...9, A B C D, & * % $)
4. **Lógico (booleano):** toda informação que pode assumir apenas duas situações (verdadeiro ou falso).

Esses tipos primitivos formam a base para **a manipulação de dados em programas de computador.** As operações sobre esses tipos, como adição, subtração, concatenação, entre outras, são frequentemente suportadas diretamente pelas linguagens de programação. **Além disso, os tipos primitivos são utilizados na declaração de variáveis.**

> Os dados matéria-prima que faz com que seja necessária a existência de computadores, pois eles são capazes de manipular e armazenar um grande volume de dados de forma rápida. Ao possibilitar a manipulação eficaz dessas informações, os computadores desempenham um papel fundamental na nossa capacidade de organizar, analisar e extrair insights valiosos a partir dos dados, impulsionando avanços significativos em diversas áreas.


## O que é uma Variável e como declarar uma em Python?

> Variável é um nome que faz referência a um valor ou objeto.

Variáveis são utilizadas para armazenar valores que queiramos utilizar em outro momento do programa. Podemos dizer também que são espaços na memória do computador que **armazenam valores que podem ser alterados durante a execução do programa.**

É como uma etiqueta que colocamos naquele valor e quando precisamos usar, chamamos pelo nome que foi atribuido a ele.

> Quando dizemos que **declaramos uma variável** significa que definimos um nome pra ela e atribuimos um valor que será armazenado nela. O operador de atribuição (sinal de igualdade **=** ) vincula um nome de variável a um valor.


Por fim, para definirmos o nome de uma variável, precisamos seguir algumas regras:

* É uma boa prática definirmos o nome da variável de acordo com o valor que ela irá armazenar, por exemplo, se for a idade de alguém, então podemos difini-la assim: **idade = 13**. Isso torna o código mais legível e compreensível.

* Deve começar com uma letra ou o caractere de sublinhado (_)

* Não pode começar com um número

* Pode conter apenas caracteres alfanuméricos e sublinhados (A-z, 0-9 e _ )

* Os nomes das variáveis diferenciam maiúsculas de minúsculas (nome, Nome e NOME são três variáveis diferentes)

* Não pode conter espaços

* Não use palavras reservadas da linguagem como nomes de variáveis. Por exemplo, em JavaScript, você não pode ter uma variável chamada function ou if.

#### Convertendo um dado numérico para booleano

Quando convertemos um inteiro ou um decimal para um booleano, o valor True será exibido quando seu valor for diferente de zero (0 ou 0.0). Caso contrário, ou seja, caso o inteiro ou o decimal seja igual a zero, o resultado será False.

````python
numFalse = 0.0
numTrue = 2
numTrue1 = -2.3

print(bool(numFalse))
print(bool(numTrue))
print(bool(numTrue1))
````

### Dados Constantes

> Entendemos que um dado é constante quando não sofre nenhuma variação no decorrer do tempo, ou seja, o valor é igual desde o início até o fim da execução do algoritmo.

Em algumas linguagens podemos definir uma variável como constante fazendo que não seja possível mudar seu valor de jeito nenhum, como em javascript que utilizamos **const idade = 12**, onde a palavra-chave CONST, própria da linguagem JS, define aquela variável como constante.

Em python não temos uma função do tipo, então qualquer variável declarada pode ser mudada ao longo do programa.

### Linguagem Tipada e Não Tipada

Nas **linguagens tipadas** ou tipagem estática, os tipos de dados são definidos durante a compilação ou antes da execução do programa. Isso significa que, ao declarar uma variável, você precisa especificar o tipo dela, e essa informação é verificada em tempo de compilação. Se tentar atribuir um valor de tipo diferente, resultará em um erro.

```
# Por exemplo, em JAVA não podemos criar uma váriavel assim idade = 12, primeiros precisamos definir o tipo de dado que essa variável irá receber:

int idade = 12 # observe que colocamos o int no começo, isso significa que essa variável ira armazenar um número inteiro
```

Nas **linguagens não tipadas**, também chamadas de linguagens de tipagem **dinâmica**, os tipos de dados são associados aos valores em tempo de execução. As variáveis podem ser usadas para armazenar diferentes tipos de dados em momentos diferentes do programa. Basicamente como fizemos em python, onde apenas atribuimos um valor sem especificar o tipo de dado que ela irá armazenar, podendo mudar seu tipo de dado de um inteiro pra um texto, por exemplo.

### Parte Inteira ou Divisão Inteira (//)

Esse operador retorna a parte inteira do resultado da divisão entre dois números, eliminando qualquer parte fracionária. Por exemplo, imagine que você queira dividir 11 por 2. O resultado é 5.5, com parte inteira 5 e parte não inteira 0.5. Dessa forma, quando utilizamos o operador de divisão inteira (//), pegamos apenas 5.

````python
x = 11
y = 2
z = x // y

print(f'A parte inteira entre a divisão de x e y é igual a {z}')
````

### Funções (built-ins) ABS, ROUND e POW

> Funções bult-ins referem-se a um conjunto de funções que são integradas diretamente na linguagem e não requerem importação de módulos específicos para serem utilizadas. Essas funções fornecem funcionalidades essenciais e comuns que podem ser utilizadas em diversos contextos (o print() é um deles em pyhton).

* **abs()** = retorna o valor absoluto.
* **round()** = retorna o valor com duas casas decimais (a quantidade você decide)
* **pow()** = função de potência



````python
num_abs = abs(8)
num_abs2 = abs(-8)

round_2_casas = round(3.14, 2)
round_3_casas = round(3.893483928498293, 3)

num = pow(2, 2) #dois elevedo a dois
num2 = pow(5, 3) #cinco elevado a 3

print(num_abs)
print(num_abs2)

print(round_2_casas)
print(round_3_casas)

print(num)
print(num2)
````

### Função JOIN

> Embora possamos concatenar strings usando o +, por questões de eficiência, deve-se usar o método **join** para concatenar um número maior de strings:



````python
a = "abelha"
b = "Borboleta"
c = "Corvo"
d = "Dinossauro"
e = "Elefante"
f = "Farofa"

print(" ".join([a, b, c, d, e, f]))

# repare que o " ". adicionou automaticamente um espaço a cada variável

print("-".join([a, b, c, d, e, f]))
````

### F-Strings

F-strings, ou formatted string literals, são uma maneira conveniente e legível de formatar strings em Python. Essas strings fornecem uma sintaxe mais simples e direta para incorporar variáveis e expressões em strings.

A principal característica das F-strings é o uso do **prefixo f** antes das aspas que definem a string. Dentro da string, você pode incluir expressões Python diretamente, indicando-as com chaves {}.

````python
nome = "Alice"
idade = 25

# F-string
mensagem = f"Olá, {nome}! Você tem {idade} anos."

print(mensagem)
````

````python
print("%i plus %i is equal to %i" % (1, 3, 4))     # inserindo as variáveis para substituir o %i respectivamente

print("{} plus {} is equal to {}".format(1, 3, 4)) # a mesma coisa do %i mas afora utilizado as chaves e a função format

print(f"{1} plus {3} is equal to {4}") # colocando as variáveis diretamente dentro das cha
````

### Verificando a presença de caracteres em uma string usando funções is

> As funções is em Python retorna True se todos os caracteres de uma string atendem a determinada condição e False caso contrário.

Essas funções podem ser usadas para validar a entrada do usuário ou para verificar se uma string contém um determinado tipo de caractere. Por exemplo, você pode usar a função isdigit() para verificar se uma string contém apenas dígitos, ou a função isalpha() para verificar se uma string contém apenas letras.

Abaixo, segue uma lista de algumas dessas funções:

* **isdigit():** Retorna True se todos os caracteres da string são dígitos, e False caso contrário.
* **isalpha():** Retorna True se todos os caracteres da string são letras do alfabeto, e False caso contrário.
* **isalnum():** Retorna True se todos os caracteres da string são alfanuméricos (letras ou dígitos), e False caso contrário.
* **islower():** Retorna True se todos os caracteres da string são letras minúsculas, e False caso contrário.
* **isupper():** Retorna True se todos os caracteres da string são letras maiúsculas, e False caso contrário.
* **isspace():** Retorna True se todos os caracteres da string são caracteres de espaço em branco, como espaço, tabulação ou quebra de linha, e False caso contrário.
